```python
import pandas as pd
```

```python
topic_df = pd.read_csv('news.csv')#, columns=['text'])
topic_df = topic_df[topic_df.rubric.isin(['Культура','Спорт', 'Экономика'])][['text', 'rubric']]
topic_df
```

```python
topic_df.rubric.value_counts()
```

```python
import spacy
nlp = spacy.load("ru_core_news_sm")
```

```python
def clean_text(headline):
  doc = nlp(headline)
  for token in doc:
      tokens = [token.lemma_ for token in doc if not token.is_stop if not token.is_punct if not token.is_digit]

  cleaned_text=" ".join(tokens)
  return cleaned_text
```

```python
topic_df['clean'] = topic_df['text'].apply(lambda x: clean_text(x))
```

```python
topic_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(topic_df['clean'])
X = X.toarray()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
target = le.fit_transform(topic_df.rubric)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model=LatentDirichletAllocation(n_components=3,evaluate_every = -1,learning_method='online', max_iter=100)
lda_top=lda_model.fit_transform(X)

# composition of doc 0 for eg
print("Document 0: ")
for i,topic in enumerate(lda_top[1000]):
  print("Topic ",i,": ",topic*100,"%")

# TFIDF нельзя использовать почему??
print(pd.DataFrame(lda_top).idxmax(axis=1).value_counts())
```

```python
target_hat = list(pd.DataFrame(lda_top).idxmax(axis=1))
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
adjusted_rand_score(target, target_hat)
```

```python
import pyLDAvis
import pyLDAvis.sklearn
import numpy as np

pyLDAvis.enable_notebook()
dtm = np.matrix(X)
panel = pyLDAvis.sklearn.prepare(lda_model, dtm, vectorizer)
panel
```

```python
# most important words for each topic
vocab = vectorizer.get_feature_names()

for i, comp in enumerate(lda_model.components_):
    vocab_comp = zip(vocab, comp)
    sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:40]
    print("Topic "+str(i)+": ")
    for t in sorted_words:
        print(t[0],end=" ")
    print("\n")
```

```python
from wordcloud import WordCloud
# Generate a word cloud image for given topic
def draw_word_cloud(index):
  imp_words_topic=""
  comp=lda_model.components_[index]
  vocab_comp = zip(vocab, comp)
  sorted_words = sorted(vocab_comp, key= lambda x:x[1], reverse=True)[:50]
  for word in sorted_words:
    imp_words_topic=imp_words_topic+" "+word[0]

  wordcloud = WordCloud(width=600, height=400).generate(imp_words_topic)
  plt.figure( figsize=(5,5))
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.tight_layout()
  plt.show()
```
